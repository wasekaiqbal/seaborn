```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('/Weather Dataset - Trial Activity DataSet.csv.csv')
df.head()
```

```python
sns.swarmplot(x='weather_type', y='wind_direction', data = df, palette ='winter')
```

```python
sns.barplot(x='weather_type', y='wind_direction', data = df, palette ='winter')
```

```python
sns.countplot(x='weather_type', data=df)
```

```python
sns.displot(df['humidity'])
```

```python
sns.pointplot(x='air_pollution_index', y='wind_speed', data=df)
```

```python
sns.pairplot(df)
```

```python
sns.jointplot(x='temperature', y='humidity',data = df, palette='winter')
```

```python
sns.lmplot(x='humidity', y='temperature',hue = 'weather_type',data = df)
```

```python
sns.stripplot(x='weather_type',y='wind_direction', data=df,palette='summer')
```
